```python

import requests

# URL صفحه خبری (می‌تونی هر سایت خبری دیگه‌ای رو تست کنی)
url = "https://www.tasnimnews.com/"

# ارسال درخواست و دریافت محتوای صفحه
response = requests.get(url)

# بررسی وضعیت پاسخ
if response.status_code == 200:
    print("✅ درخواست موفق بود!")
    print(response.text[:500])  # نمایش 500 کاراکتر اول HTML صفحه
else:
    print("❌ خطا در دریافت صفحه:", response.status_code)
```

```python
from bs4 import BeautifulSoup

# پردازش محتوای HTML
soup = BeautifulSoup(response.text, "html.parser")

# پیدا کردن همه تگ‌های مربوط به عنوان خبرها
titles = soup.find_all("h2")  # بستگی به سایت، ممکنه h2، h3 یا a باشه

# نمایش عنوان‌ها
print("📢 عناوین اخبار:")
for title in titles[:10]:  # نمایش 10 عنوان اول
    print("-", title.text.strip())
```

```python
# پیدا کردن لینک‌های خبرها
news_links = soup.find_all("a")  # پیدا کردن همه تگ‌های <a> (لینک‌ها)

print("📰 لیست اخبار:")
for link in news_links[:10]:  # نمایش 10 خبر اول
    title = link.text.strip()  # گرفتن متن لینک
    href = link.get("href")  # گرفتن آدرس لینک
    if title and href:  # اگر عنوان و لینک وجود داشته باشه
        print(f"- {title}: {href}")
```

```python
import csv

# باز کردن یا ایجاد فایل CSV برای ذخیره داده‌ها
with open("news.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["عنوان خبر", "لینک خبر"])  # نوشتن هدر فایل

    # نوشتن اطلاعات هر خبر در فایل
    for link in news_links[:10]:
        title = link.text.strip()
        href = link.get("href")
        if title and href:
            writer.writerow([title, href])

print("✅ داده‌ها در فایل 'news.csv' ذخیره شدند!")
```
